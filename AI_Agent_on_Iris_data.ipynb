## Data visualization

### Subtask:
Create a scatter plot to visualize the relationship between sepal length and sepal width.


**Reasoning**:
Create a scatter plot to visualize the relationship between sepal length and sepal width using matplotlib.pyplot, including labels, title, and potentially different colors based on species.



```python
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df_iris['SepalLengthCm'], df_iris['SepalWidthCm'], c=df_iris['Species'].astype('category').cat.codes)

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Relationship between Sepal Length and Width')

# Add a legend (if needed, based on how you color the points)
# You may need to adjust this part depending on how you set the colors
# plt.legend()

plt.show()
```

**Reasoning**:
The previous code block generated a plot, but the legend is missing. We can add a legend by creating a unique color for each species and adding a corresponding label to the legend.



```python
import matplotlib.pyplot as plt

# Create the scatter plot with different colors for each species
plt.figure(figsize=(8, 6))
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}
for species, color in colors.items():
    plt.scatter(df_iris[df_iris['Species'] == species]['SepalLengthCm'],
                df_iris[df_iris['Species'] == species]['SepalWidthCm'],
                color=color, label=species)

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Relationship between Sepal Length and Width')

# Add a legend
plt.legend()

plt.show()
```
