```python
# Predictive analytics using machine learning
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = {
    'Sales': [100, 120, 110, 130, 140, 150, 160, 170, 180, 190],
    'Marketing_Spend': [10, 12, 11, 13, 14, 15, 16, 17, 18, 19]
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(df[['Marketing_Spend']], df['Sales'], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)
```

```python
# Supply chain optimization using optimization algorithms
import numpy as np
from scipy.optimize import minimize

demand = np.array([100, 120, 110, 130, 140, 150, 160, 170, 180, 190])
supply = np.array([10, 12, 11, 13, 14, 15, 16, 17, 18, 19])

def objective(x):
    return np.sum((x - demand) ** 2)

bounds = [(0, None) for _ in range(len(supply))]

res = minimize(objective, supply, method='SLSQP', bounds=bounds)

print(res.x)
```

```python
# Inventory management using simulation
import numpy as np
import matplotlib.pyplot as plt

demand = np.array([100, 120, 110, 130, 140, 150, 160, 170, 180, 190])
lead_time = 2
safety_stock = 10

inventory = 100

inventory_levels = []
for i in range(len(demand)):
    net_demand = demand[i] - inventory

    if net_demand > 0:
        inventory += net_demand + safety_stock

    inventory_levels.append(inventory)

    if i >= lead_time:
        inventory -= demand[i - lead_time]

plt.plot(inventory_levels)
plt.xlabel('Time')
plt.ylabel('Inventory Level')
plt.show()
```

```python
# Demand forecasting using time series analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

data = {
    'Date': pd.date_range('2022-01-01', periods=12),
    'Sales': [100, 120, 110, 130, 140, 150, 160, 170, 180, 190, 200, 210]
}

df = pd.DataFrame(data)

model = ARIMA(df['Sales'], order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=3)

plt.plot(df['Sales'])
plt.plot(forecast)
plt.xlabel('Time')
plt.ylabel('Sales')
plt.show()
```
